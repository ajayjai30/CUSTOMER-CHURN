```python
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,mean_squared_error,mean_absolute_error,r2_score
```

```python
#LinearRegression
x_train,x_test,y_train,y_test=train_test_split(df['tenure'],df['MonthlyCharges'],test_size=0.3,random_state=1)
```

```python
x_train=pd.DataFrame(x_train)
y_train=pd.DataFrame(y_train)
x_test=pd.DataFrame(x_test)
y_test=pd.DataFrame(y_test)
```

```python
f=LinearRegression()
f.fit(x_train,y_train)
```

```python
y_pred=f.predict(x_test)
```

```python
error1=mean_squared_error(y_pred,y_test)
error2=mean_absolute_error(y_pred,y_test)
error3=r2_score(y_pred,y_test)
print("Mean squared error",error1)
print("Mean absolute error",error2)
print("r2_score",error3)
```

```python
print("RMSE for linear regression:",np.sqrt(mean_squared_error(y_pred,y_test)))
```
